```python
import torch

def count_unique_objects(results, iou_threshold=0.5):
    """
    Counts unique objects by applying IoU-based filtering.
    
    :param results: YOLO model inference results
    :param iou_threshold: Overlap threshold for removing duplicate detections
    :return: Unique apple count
    """
    import torchvision.ops as ops  # PyTorch IoU function

    boxes = results.xyxy[0]  # Bounding box coordinates
    scores = results.pandas().xyxy[0]["confidence"].values  # Confidence scores

    if len(boxes) == 0:
        return 0  # No apples detected

    # Convert to tensor format required for IoU calculation
    boxes_tensor = torch.tensor(boxes)

    # Apply Non-Maximum Suppression (NMS) to remove duplicate apples
    keep_indices = ops.nms(boxes_tensor, torch.tensor(scores), iou_threshold)
    unique_count = len(keep_indices)

    return unique_count

try:
    # Load YOLO model
    model = torch.load(r'C:\Users\Rushikesh\Desktop\Fruit_counting - Copy\best.pt')

    # Input image
    imgs = [r'C:\Users\Rushikesh\Desktop\Fruit_counting - Copy\Apple_Screenshot 2025-03-17 175852.png']

    # Perform inference
    results = model(imgs)

    # Count unique apples
    num_apples = count_unique_objects(results)

    print(f"Total Unique Apples Detected: {num_apples}")

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import torch
import torchvision.ops as ops  # IoU-based filtering

# Load the YOLO model correctly
from ultralytics import YOLO  # Ensure you have YOLOv8 installed (`pip install ultralytics`)

def count_unique_objects(results, iou_threshold=0.5):
    """
    Counts unique apples by applying IoU-based filtering.
    :param results: YOLO model inference results
    :param iou_threshold: Overlap threshold for duplicate removal
    :return: Unique apple count
    """
    boxes = results[0].boxes.xyxy.cpu()  # Get bounding boxes
    scores = results[0].boxes.conf.cpu()  # Get confidence scores

    if len(boxes) == 0:
        return 0  # No apples detected

    # Apply Non-Maximum Suppression (NMS) to remove duplicate apples
    keep_indices = ops.nms(boxes, scores, iou_threshold)
    return len(keep_indices)

try:
    # Load YOLOv8 model (ensure correct path)
    model = YOLO(r'C:\Users\Rushikesh\Desktop\Fruit_counting - Copy\best.pt')  

    # Perform inference
    results = model(r'C:\Users\Rushikesh\Desktop\Fruit_counting - Copy\Apple_Screenshot 2025-03-17 175852.png')  # <---- FIXED THIS LINE

    # Count unique apples
    num_apples = count_unique_objects(results)

    print(f"Total Unique Apples Detected: {num_apples}")

except Exception as e:
    print(f"An error occurred: {e}")
```
